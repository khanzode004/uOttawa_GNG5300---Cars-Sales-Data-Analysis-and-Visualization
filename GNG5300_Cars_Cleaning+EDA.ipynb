## Data Cleaning and Preparation

```python
#importing all the required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the dataset csv file into a Pandas Dataframe
df = pd.read_csv('data.csv')
```

```python
#description of the dataset
df.describe()
```

```python
#dataset information
df.info()
```

```python
#dataset information
df.info()
```

```python
#first 5 rows of the dataset
df.head()
```

```python
#last 5 rows of the dataset
df.tail()
```

```python
#changing some column names
df.rename(columns={'Driven_Wheels': 'Driving Wheels', 'highway MPG':'Highway Mileage (mpg)', 'city mpg':'City Mileage (mpg)', 'Popularity':'Popularity Rating'}, inplace=True)
```

```python
#current shape of the dataset
df.shape
```

```python
#listing out data types of each feature
df.dtypes
```

```python
#counting non null values in each column
df.count()
```

```python
#dropping any duplicate rows
df = df.drop_duplicates()
```

```python
#current shape of the dataset
df.shape
```

```python
#changing the null values in the 'Market Category' feature to 'Miscellaneous' instead of dropping the rows all together.
df['Market Category'] = df['Market Category'].fillna('Miscellaneous')
```

```python
#total non-null values in each column
print(df.isnull().sum())
```

```python
#dropping any rows which have 
df = df.dropna()
```

```python
#changing the features to more appropriate datatypes.
df['Year'] = df['Year'].astype(str)
df['Engine Cylinders'] = df['Engine Cylinders'].astype(int)
df['Number of Doors'] = df['Number of Doors'].astype(int)
df['Highway Mileage (mpg)'] = df['Highway Mileage (mpg)'].astype(float)
df['City Mileage (mpg)'] = df['City Mileage (mpg)'].astype(float)
df['Popularity Rating'] = df['Popularity Rating'].astype(float)
df['MSRP'] = df['MSRP'].astype(float)
```

```python
#listing out data types of each feature
df.dtypes
```

```python
#total non-null values in each column remaining
print(df.isnull().sum())
```

```python
#current shape of the dataset
df.shape
```

```python
df['Vehicle Size'].value_counts()
```

```python
#resetting indexing of the dataset after dropping duplicates and null-values
df.reset_index(drop=True)
```

## Grouped Statistics 

### Engine Cylinders v/s Engine HP 

```python
#plotting boxplots to show relationship between number of engine cylinders and engine HP
plt.figure(figsize=(12,12))
sns.boxplot(data=df, x='Engine Cylinders', y='Engine HP')
plt.show()
```

### Fuel Type v/s MSRP

```python
#grouping up data by fuel types and calculating mean MSRP for each
fuel_group_prices = df.groupby('Fuel Type Group')['MSRP'].mean().reset_index()
fuel_prices = df.groupby('Engine Fuel Type')['MSRP'].mean().reset_index()
```

```python
#plotting bar chart which shows relationship between Fuel Type and MSRP
plt.figure(figsize=(12,8))
sns.barplot(data=fuel_group_prices, y='Fuel Type Group', x='MSRP')
plt.show()
```

### Make v/s MSRP 

```python
#grouping up data by Make and calculating mean MSPR for each
bymake = df.groupby('Make')['MSRP'].mean().reset_index()
bymake.sort_values(by='MSRP', ascending=False, inplace=True)
```

```python
#plotting average MSRP for each make (under 150K)
plt.figure(figsize=(12,16))
sns.barplot(data=bymake[bymake.MSRP < 0.15e6], y='Make', x='MSRP')
plt.xlim([0, 0.15e6])
plt.ylabel(None)
plt.xlabel('Manufacturer Suggested Retail Price')
plt.title("Average Price of Car by Make (under $150000)")
plt.show()
```

```python
#plotting average MSRP for each make (over 150K)
plt.figure(figsize=(12,8))
sns.barplot(data=bymake[bymake.MSRP > 0.15e6], y='Make', x='MSRP')
plt.ylabel(None)
plt.xlabel('Manufacturer Suggested Retail Price')
plt.title("Average Price of Car by Make (over $150000)")
plt.show()
```

### Year v/s MSRP 

```python
# grouping up data by year and plotting a bar graph for average MSRP values for each
byyear = df.groupby('Year')['MSRP'].mean().reset_index()
plt.figure(figsize=(12,16))
sns.barplot(data=byyear[byyear.MSRP < 0.15e6], y='Year', x='MSRP')
plt.ylabel(None)
plt.xlabel('Manufacterer Suggested Retail Price')
plt.title("Average Price by Year")
plt.show()
```
